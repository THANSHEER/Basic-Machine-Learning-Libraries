# **Difference Between AI and ML**

As already mentioned, Machine learning is a field that arises from Artificial Intelligence. **Artificial Intelligence** helps in building systems by **explicitly programming them for different tasks**, on the other hand, **Machine Learning** focuses on completing a single task successfully **by providing large amounts of data to the machine for learning on its own** and not programming it explicitly.

# **Introduction to Supervised Learning, Unsupervised Learning and Reinforced learning**

### **Supervised Learning**

One of the main techniques used for building AI/ML powered systems is **Supervised Learning**. In this method, huge amount of labelled dataset is used. With the help of this, the system learns on its own by identifying different features, classes in the dataset. The results generated by system are evaluated by a superviser and tells the system whether the results are correct or not. 

**Example** - Facebook is able to identify people in the photos uploaded on it by using the same Supervised Learning technique.

Facebook_Photo_Taggin.jpg

**Training Set**

A labelled dataset contains different classes for prediction. Dataset is divided into two parts. The first part is training set that helps in training the system.  

**Testing Set**

Testing set is the other part that helps in evaluating our system's performance. we can try to predict the results on this testing set by using the training set knowledge.

### **Unsupervised Learning**

**Unsupervised Learning** is another majorly used technique for building AI/ML enabled systems. In this case, there is no training data, the approach used here is to identify patterns in the data available and group the different elements into different categories on the basis of similarities.

**Example** - Grouping of different news articles by Google on the basis of topics these articles cover.

Unsupervised_Learning.png

# **What is Classification?**

**Classification** is a process of finding a function which helps in dividing the dataset into classes based on different parameters. In Classification, a computer program is trained on the training dataset and based on that training, it categorizes the data into different classes.

The task of the classification algorithm is to find the mapping function to map the input(x) to the discrete output(y).

**Example**: The **best example** to understand the Classification problem is **Email Spam Detection**. The model is trained on the basis of millions of emails on different parameters, and whenever it receives a new email, it identifies whether the email is spam or not. If the email is spam, then it is moved to the Spam folder.

There are mainly **two types of classification**, these two are depicted with the help of following visualization

Types_of_Classification.png

The classification process deal with the problems where the data can be divided into binary or multiple discrete labels.

**Let's look at the Types of ML Classification Algorithms**:

Classification Algorithms can be further divided into the following types:

* Logistic Regression
* K-Nearest Neighbours
* Support Vector Machines
* Kernel SVM
* Naïve Bayes
* Decision Tree Classification
* Random Forest Classification


# **What is Regression?**

**Regression** is a process of finding the correlations between dependent and independent variables. It helps in **predicting the continuous variables** such as prediction of Market Trends, prediction of House prices, etc.

The task of the Regression algorithm is to find the mapping function to map the input variable(x) to the continuous output variable(y).

**Example**: Suppose we want to do weather forecasting, so for this, we will use the Regression algorithm. In weather prediction, the model is trained on the past data, and once the training is completed, it can easily predict the weather for future days.
